```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6,7,8,9]
y2=[2,4,5,7,6,8,9,6]
y3=[9,8,7,6,5,4,3,2]
plt.scatter(x,y2,label="y2",color="grey",
         marker="*",s=30)
plt.scatter(x,y3,label="y3",color="red",
         marker="o",s=30)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('my scatter plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
activities=['eat','sleep','work','play']
slices=[3,7,8,6]
colors=['r','y','g','b']
plt.pie(slices,labels=activities,colors=colors,
       startangle=90,shadow=True,
       explode=(0,0,0.2,0),radius=2.2,
       autopct='%2.if%%')
plt.legend()
plt.show()
```
